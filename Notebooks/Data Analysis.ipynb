```python
#import libraries
import pandas as pd
import seaborn as sns
```

```python
#read in NFL combine  file
file = '..\\Data\\general_data.csv'
df = pd.read_csv(file, index_col=8)
df.head()
```

```python
df.describe(include='all')
```

```python
#look at datatypes
df.dtypes
```

```python
#check to see if there are any null values
df.isnull().sum()
```

```python
#drop rows with na value
df.dropna(inplace=True)
```

```python
#add new gender column
df['Gender_Code'] = 0
df.loc[df['Gender'] == 'Female', 'Gender_Code'] = 1
```

```python
df.head()
```

```python
#see if there are any outliers for monthly 
sns.boxplot(data=df, x='MonthlyIncome')
```

```python
#see if there are any outliers for working years 
sns.boxplot(data=df, x='TotalWorkingYears')
```

```python
#save new dataset to csv
file = '..\\Data\\Final_Data.csv'
df.to_csv(file)
```
