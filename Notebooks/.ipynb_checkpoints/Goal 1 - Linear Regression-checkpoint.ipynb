```python
#import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from Functions import *
```

```python
#read in final dataset
file = '..\Data\Final_Data.csv'
df = pd.read_csv(file, index_col=0)
df.head()
```

```python
#look at quick plot
sns.lmplot(x = 'TotalWorkingYears', y = 'YearsAtCompany', data = df, ci = 0)
plt.draw()
```

```python
#create linear regression
X = df[['TotalWorkingYears']]
y = df.YearsAtCompany
crossval = lin_reg(X, y, cv=10)
crossval
```

```python
#pick linear regression, add new column, and display 
estimator = crossval['estimator'][9]
pred = estimator.predict(df[['TotalWorkingYears']])
df['Pred_1'] = pred
df[['TotalWorkingYears', 'YearsAtCompany', 'Pred_1']]
```

```python
#create another linear regression
X = df[['MonthlyIncome', 'YearsSinceLastPromotion', 'TrainingTimesLastYear', 'YearsWithCurrManager']]
y = df.YearsAtCompany
crossval = lin_reg(X, y, cv=10)
crossval
```

```python
#use the 10th iteration of the linear regression as that it has the best test r2 score and comparable train score
estimator = crossval['estimator'][2]
pred = estimator.predict( df[['MonthlyIncome', 'YearsSinceLastPromotion', 'TrainingTimesLastYear', 'YearsWithCurrManager']])
df['Pred_2'] = pred
df[['MonthlyIncome', 'YearsSinceLastPromotion', 'TrainingTimesLastYear', 'YearsWithCurrManager', 'Pred_2']]
```
