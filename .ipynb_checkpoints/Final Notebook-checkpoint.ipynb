## Goal 1

Can we use a linear regression to predict how long someone someone has worked at the company?

We will use TotalWorkingYears, MonthlyIncome, YearsSinceLastPromotion, TrainingTimesLastYear, YearsWithCurrManager as input and YearsAtCompany as output.  Upon the creation of a successful this linear regression can be used to determine what makes an employee stay and begin looking into those attributes.

```python
#create linear regression
X = df[['MonthlyIncome', 'YearsSinceLastPromotion', 'TrainingTimesLastYear', 'YearsWithCurrManager']]
y = df.YearsAtCompany
crossval = lin_reg(X, y, cv=10)
crossval
```

```python
#use the 10th iteration of the linear regression as that it has the best test r2 score and comparable train score
estimator = crossval['estimator'][2]
pred = estimator.predict( df[['MonthlyIncome', 'YearsSinceLastPromotion', 'TrainingTimesLastYear', 'YearsWithCurrManager']])
df['Prediction'] = pred
```

```python
#display predictions along with actual data and input featured
df[['YearsAtCompany', 'Prediction', 'MonthlyIncome', 'YearsSinceLastPromotion', 'TrainingTimesLastYear', 'YearsWithCurrManager']]
```

## Results

The linear regression chosen had a test r-squared of approx. 68% and a train of approx. 66%.  This is a decent linear regression.  Since we can make some predictions with this linear regression, we can then determine what needs to be needs to happen in order for new and existing employees need to be retained, ex. additional training, a certain amount of years with a manager, etc.

```python
#determine kmeans groups
kmeans_groups(min=2, max=10, df = df[['TotalWorkingYears', 'MonthlyIncome', 'YearsSinceLastPromotion', 'TrainingTimesLastYear', 'YearsWithCurrManager']])
```
